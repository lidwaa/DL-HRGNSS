# *Time series examples*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
idx = 1 # index in dataset
dt = 181 # time serie length [181 or 501]
ns = 3 # stations [3 or 7]
dirdata = './data/GNSS_M'+str(ns)+'S_'+str(dt)+'/'
xtest = np.load(dirdata+'xdata.npy')
ytest = np.load(dirdata+'ydata.npy')
```

```python
plt.rcParams.update({'font.size': 14, 'axes.labelpad':12})
fig, ax = plt.subplots(ns,1,figsize=(8,ns*3),sharex=True,gridspec_kw={'hspace': 0})
ax[0].set_xlim(0,dt)
for i in range(ns):
    ax[i].plot(xtest[idx][:,:][i],label = ['Z','N','E'])
    ax[i].legend(loc='upper right',fontsize=12)
    x = np.amax(xtest[idx][:,:][i], axis=0)
    rx = max(x)
    ax[i].text(5,rx-(rx/4),'station '+str(i+1),fontweight='bold')

ax[ns-1].set_xlabel('Time [sec]')
ax[round(ns/2)-1].set_ylabel('Displacement [m]')
plt.show()
```
